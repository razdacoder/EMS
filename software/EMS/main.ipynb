### Automated Examination Timetabling System using a Brute Force Approach

```python
import pandas as pd
```

```python
dates = pd.date_range(start="2023-02-01", end="2023-02-14").to_pydatetime().tolist()
dates
```

```python
courses = pd.read_csv('courses.csv', names=["Code", "Title"])
```

```python
courses.head()
```

```python
def apply_dep(value: str) -> str:
    if value.startswith("SS"):
        return "general"
    elif value.startswith("CS") or value.startswith("AI"):
        return "CS"
    else:
        return "MS"
```

```python
courses["Department"] = courses["Code"].apply(apply_dep)
```

```python
courses.head()
```

```python
courses["Department"].unique()
```

```python
courses.head(20)
```

```python
cs_courses = courses[courses["Department"] != "MS"]["Code"]
cs_course = cs_courses.to_list()
```

```python
cs = []
for c in cs_course:
    cs.append({"Code": c, "Department": "CS"})
```

```python
ms_courses = courses[courses["Department"] != "CS"]["Code"]
ms_course = ms_courses.to_list()
```

```python
ms = []
for c in ms_course:
    ms.append({"Code": c, "Department": "MS"})
```

```python
ms , cs
```

```python
classes = [cs, ms]
```

```python
from random import shuffle
```

```python
schedule = []

def check_course(code):
    if len(schedule) == 0:
        return False
    for sc in schedule:
        if sc["Course"] == code:
            return True
        else:
            False


def fix_break(class_courses):
    if len(class_courses) == len(dates):
        return dates
    elif len(class_courses) < len(dates):
        num_breaks = len(dates) - len(class_courses)
        for i in range(num_breaks):
            
    else:
        pass
    

# def get_date(code):
#     for sc in schedule:
#         if sc["Course"] == code:
#             return sc["Date"]


# def check_date(code, date):
#     for sc in schedule:
#         if sc["Course"] == code and sc["Date"] == date:
#             return True
#         else:
#             return False



for class_c in classes:
    shuffle(class_c)
    date_index = 0
    for i in range(0, len(class_c)):
        if(check_course(class_c[i]["Code"])):
            print(f"{class_c[i]['Code']} Scheduled Already!!")
        else:
            sc = {
                "Course": class_c[i]["Code"],
                "Department": class_c[i]["Department"],
                "Date": dates[date_index]
            }
            date_index += 1
            schedule.append(sc)
    
            
schedule
```

```python
schedules_df = pd.DataFrame(schedule)
```

```python
schedules_df.to_csv("schedule.csv")
```
